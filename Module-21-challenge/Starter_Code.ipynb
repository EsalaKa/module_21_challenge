## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns 'EIN' and 'NAME'
application_df = application_df.drop(['EIN', 'NAME'], axis=1)


```

```python
# Determine the number of unique values in each column
unique_counts = application_df.nunique()


print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Choose a cutoff value for replacing 
cutoff_value = 500  

# Create a list of application types to be replaced with "Other"
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()

# Replace in the dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value
cutoff_value = 100

# Find application types that occur less than or equal to the cutoff value
infrequent_application_types = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() <= cutoff_value].index.tolist()

# Create a list of application types to be replaced with "Other"
application_types_to_replace = infrequent_application_types

# Replace in the DataFrame
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check the value counts after binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts where count is greater than 1
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts_gt1 = classification_counts[classification_counts > 1]
print(classification_counts_gt1)
```

```python
# Look at CLASSIFICATION value counts
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Choose a cutoff value for replacing less frequent classifications
cutoff_value = 100  

# Create a list of classifications to be replaced with "Other"
classifications_to_replace = classification_counts[classification_counts <= cutoff_value].index.tolist()

# Replace in the dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
classification_counts_after_binning = application_df['CLASSIFICATION'].value_counts()
print(classification_counts_after_binning)
```

```python
# List of categorical columns to be converted to dummies
categorical_columns = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']

# Apply pd.get_dummies to each categorical column
application_df_encoded = pd.get_dummies(application_df, columns=categorical_columns)

# Display the resulting DataFrame
application_df_encoded.head()
```

```python

X = application_df_encoded.drop(columns=['IS_SUCCESSFUL'])
y = application_df_encoded['IS_SUCCESSFUL']

# Import the train_test_split function
from sklearn.model_selection import train_test_split

# Split the data, with training set (70%) and a testing set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
import tensorflow as tf

# Define the model 
nn = tf.keras.models.Sequential()

# Input layer with theinput features 
input_features = X_train.shape[1]

# First hidden layer with 32 nodes and ReLU activation function
nn.add(tf.keras.layers.Dense(units=32, activation='relu', input_dim=input_features))

# Second hidden layer with 16 nodes and ReLU activation function
nn.add(tf.keras.layers.Dense(units=16, activation='relu'))

# Output layer with a single node and sigmoid activation function 
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model_history = nn.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
